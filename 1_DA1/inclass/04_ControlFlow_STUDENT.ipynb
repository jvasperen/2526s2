## Conditions: if/elif/else

Make decisions in code!

**Comparison operators:**
- `==` equal
- `!=` not equal
- `>` greater than
- `<` less than
- `>=` greater or equal
- `<=` less or equal

```python
# Simple if
followers = 1500

if followers > 1000:
    print('Popular account!')
```

```python
# if-else
age = 22

if age >= 18:
    print('Adult')
else:
    print('Minor')
```

```python
# if-elif-else (multiple conditions)
likes = 75

if likes >= 100:
    print('Viral!')
elif likes >= 50:
    print('Good engagement')
else:
    print('Low engagement')
```

### Combining Conditions

- `and` - both must be true
- `or` - at least one must be true
- `not` - opposite

```python
likes = 450
comments = 30

if likes > 400 and comments > 20:
    print('High engagement!')
```

## For Loops: Repeat for Each Item

Loop through lists automatically!

```python
# Loop through list
platforms = ['Instagram', 'TikTok', 'Twitter']

for platform in platforms:
    print('I use', platform)
```

```python
# Loop with numbers
for i in range(5):
    print('Count:', i)
```

```python
# Combine loop + condition
scores = [7, 8, 5, 9, 6]

for score in scores:
    if score >= 7:
        print(score, '- Satisfied')
    else:
        print(score, '- Unsatisfied')
```

## While Loops: Repeat Until Condition

Keep going while something is true

```python
# Count to 5
count = 1

while count <= 5:
    print('Count:', count)
    count += 1

print('Done!')
```

## Functions: Reusable Code

Package code to use multiple times

```python
# Define function
def calculate_engagement(likes, followers):
    rate = likes / followers
    return rate
```

```python
# Use function multiple times
result1 = calculate_engagement(450, 1500)
result2 = calculate_engagement(200, 800)
print('Rate 1:', result1)
print('Rate 2:', result2)
```

```python
# Function with condition
def categorize_post(likes):
    if likes > 1000:
        return 'Viral'
    elif likes > 100:
        return 'Popular'
    else:
        return 'Normal'

print(categorize_post(1500))
print(categorize_post(250))
```
