## Part 1: Conditions (if/elif/else)

**Conditions** let your code make decisions.

Real-world example:
* **IF** it's raining, take an umbrella
* **ELSE** wear sunglasses

In Python:
```python
if followers > 1000:
    print('Popular account!')
else:
    print('Growing account')
```

### Comparison Operators

* `==` equal to
* `!=` not equal to
* `>` greater than
* `<` less than
* `>=` greater than or equal
* `<=` less than or equal

````python
# Simple if statement
age = 22

if age >= 18:
    print('Adult')
````

````python
# if-else
followers = 1500

if followers > 1000:
    print('You have many followers!')
else:
    print('Keep growing your audience')
````

````python
# if-elif-else (multiple conditions)
likes = 75

if likes >= 100:
    print('Viral post!')
elif likes >= 50:
    print('Good engagement')
elif likes >= 10:
    print('Some engagement')
else:
    print('Low engagement')
````

âš ï¸ **Important:** Indentation matters! Everything indented belongs to that condition.

### Combining Conditions

Use `and`, `or`, `not` to combine conditions:

````python
# AND - both must be true
age = 22
has_account = True

if age >= 18 and has_account:
    print('Can post content')
````

````python
# OR - at least one must be true
platform = 'Instagram'

if platform == 'Instagram' or platform == 'TikTok':
    print('Visual platform')
````

````python
# NOT - opposite
is_verified = False

if not is_verified:
    print('Account needs verification')
````

## Part 2: For Loops

**Loops** let you repeat actions automatically.

**For loops** repeat for each item in a collection:

````python
# Loop through a list
platforms = ['Instagram', 'TikTok', 'Twitter']

for platform in platforms:
    print('I use', platform)
````

````python
# Loop through numbers
for i in range(5):
    print('Count:', i)
````

````python
# range(start, stop, step)
for i in range(1, 11, 2):
    print(i)
````

````python
# Loop through a string
for letter in 'Python':
    print(letter)
````

## Part 3: While Loops

**While loops** repeat as long as a condition is True:

````python
# Count to 5
count = 1

while count <= 5:
    print('Count:', count)
    count += 1

print('Done!')
````

````python
# Growth simulation
followers = 100
weeks = 0

while followers < 1000:
    followers += 50  # Gain 50 followers per week
    weeks += 1
    print(f'Week {weeks}: {followers} followers')

print(f'Reached 1000+ followers in {weeks} weeks!')
````

âš ï¸ **Warning:** Make sure your while loop will eventually stop! Otherwise it runs forever.

## Part 4: Introduction to Functions

**Functions** package code you want to reuse.

Format:
```python
def function_name(parameters):
    # code
    return result
```

````python
# Simple function
def greet(name):
    print('Hello,', name)

greet('Emma')
greet('Noah')
````

````python
# Function with return
def calculate_engagement(likes, followers):
    rate = likes / followers
    return rate

result = calculate_engagement(450, 1500)
print('Engagement rate:', result)
````

````python
# Function with condition
def categorize_post(likes):
    if likes > 1000:
        return 'Viral'
    elif likes > 100:
        return 'Popular'
    else:
        return 'Normal'

print(categorize_post(1500))
print(categorize_post(250))
print(categorize_post(45))
````
