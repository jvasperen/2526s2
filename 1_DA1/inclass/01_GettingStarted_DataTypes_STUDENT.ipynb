## Part 5: Text in Python (Strings)

Text in Python is called a **string**. Strings must be wrapped in quotes.

You can use:
* Single quotes: `'Hello'`
* Double quotes: `"Hello"`

Both work the same way!

```python
# A string with single quotes
'This is text'
```

```python
# A string with double quotes
"This is also text"
```

```python
# Check the type
type('Hello')
```

### ‚ö†Ô∏è Common mistakes with strings

What happens if you forget the quotes?

```python
# This will cause an error - try running it!
Hello World
```

‚ùå **Error!** Python thinks `Hello` and `World` are variable names, not text.

**Always use quotes for text!**
