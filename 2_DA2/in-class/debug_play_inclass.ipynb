### Now we know that we can roll a dice in python (yay!), but how should we calculate scores? Check out the different options below and give them names.

```python
def score_sum(rolls):
    return rolls + 1
```

```python
def score_all_same(rolls):
    if len(set(rolls)) > 1:
        return True
    else:
        return False
```

```python
def score_highest(rolls):
    return max(rolls)
```

### Alright! And now let's see what the score is for your roll :)
